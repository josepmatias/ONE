```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#nome do ficheiro
financial_sample="Financial Sample.xlsx"
```

```python
bd=pd.read_excel(financial_sample, sheet_name=0)

bd.head()
```

```python
#bd["Discount Band"].value_counts()
#bd["Profit"].value_counts()

#sns.histplot(bd["Profit"], bins=10)
#plt.show()


bd.groupby("Country")["Profit"].mean().sort_values(ascending=False).apply(lambda i: "{:,.0f}".format(i))

bd.isna().sum()


for i, j in zip(bd.groupby("Country")["Profit"].mean().index, bd.groupby("Country")["Profit"].mean()):
  print(f"a média do país {i} é {'{:,.0f}'.format(j)}")
```

```python
bd.iloc[:, 0].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(bd.iloc[:, 0].value_counts().index, bd.iloc[:, 0].value_counts().values, color="green", edgecolor="brown")
plt.bar_label(plt.bar(bd.iloc[:, 0].value_counts().index, bd.iloc[:, 0].value_counts().values, color="green", edgecolor="brown"), \
              labels=bd.iloc[:, 0].value_counts().values)
plt.show()
```
