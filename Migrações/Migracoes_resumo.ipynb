```python
#ler ficheiro migracoes_mes - folha "retalho_migracoes"

retalho_migracoes=pd.read_excel(migracoes_mes_nome, sheet_name=3)

retalho_migracoes.head()
```

```python
#função para criar lista com ids de contrato iguais na folha resumo e na folha de fecho de migrações

def cruzamento (db, db_fecho, col_fecho, col_mig):
  db_fecho2=db_fecho.sort_values(by=col_fecho, ascending=True)
  db_cruzamento=db_fecho2[[col_fecho]].merge(db[[col_mig]], how="inner", left_on=col_fecho, right_on=col_mig)
  return db_cruzamento[col_fecho].tolist()

lista_cruzamento=cruzamento(resumo, retalho_migracoes, "COD_LOC_SA", "id de contrato")

```

```python
#função para actualizar a folha resumo com os novos valores para as colunas do relatório de migrações fecho



lista_cols_fecho=["COD_AGEN", "COD_DELEG_AGEN", "COD_ENT_AGEN", "COD_DIA0", "Nível 2"]
lista_cols_mig=["Cav Marcação(Retalho)", "Delegação", "Entidade", "Data Marcação", "Canal"]

retalho_migracoes.sort_values("COD_LOC_SA", ascending=True, inplace=True)
for i, j in zip(lista_cols_fecho, lista_cols_mig):
  lista_valores_fecho=retalho_migracoes.loc[retalho_migracoes["COD_LOC_SA"].isin(lista_cruzamento), i].values.tolist()

  resumo.loc[resumo["id de contrato"].isin(lista_cruzamento), j]=lista_valores_fecho
```
